# missing values and imputing them

```python
# binings of age column into 7 categories
bins=[0,1,5,12,18,30,50,80]
labels=['infants','toddlers','kids','teens','youngs','Middle-aged','Old']
```

```python
# feature engineering Using bins methods 
# binings of age column into 7 categories
bins=[0,1,5,12,18,30,50,80]
labels=['infants','toddlers','kids','teens','youngs','Middle-aged','Old']
df['binned-age']=pd.cut(df['age'],bins=bins,labels=labels)
```
