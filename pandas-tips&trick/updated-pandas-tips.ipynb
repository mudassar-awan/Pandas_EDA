### 9. üî¢ Converting String to numbers
Transforming text-based numbers into actual numeric types for calculations.

```python
df=pd.DataFrame({'col_A':['1.1','2','3','4','5'],
                 'col_B':['6','7','8','9','10']})
df
df.dtypes
```

```python
# It converts the data types of specific columns and then shows the updated data types of all columns.
df.astype({'col_A':'float64' ,'col_B':'int64'}).dtypes
```

```python
# üëâ It converts the column values to numbers.
# üëâ If a value cannot be converted, it becomes NaN instead of throwing an error.
#In simple words:
#‚ÄúTry to turn everything into numbers. If something is dirty, don‚Äôt crash ‚Äî mark it as missing.‚Äù
pd.to_numeric(df['col_A'],errors='coerce')
pd.to_numeric(df['col_B'],errors='coerce')
```

### 16. ‚úÇÔ∏è Splitting a string into multiple columns
Extracting specific parts of a string into separate, individual columns.

```python
import pandas as pd
df=pd.DataFrame({'name':["Malik hamid",'Razaq ahmed','Mosi khan','john smith'],
                 'location':['muqam kshmir','kail neelum','mzd kashmir','lahore pakistan']})
df.name.str.split(' ').head
```

```python
df.name.str.split(' ',expand=True) # splits the 'name' column by spaces into separate columns
```

```python
# splits the 'name' column into 'first-name' and 'last-name'
df[['first-name','last-name']]=df.name.str.split(' ',expand=True)
# splits the 'location' column into 'city-name' and 'country-name'
df[['city-name','country-name']]=df.location.str.split(' ',expand=True)
# selects and reorders only the new split columns for the final DataFrame
df=df[['first-name','last-name','city-name','country-name']]
df
```

### 19. üìê Reshaping of multi index series
Changing the layout of your data to make comparison between groups easier.

```python
# calculates the average of the 'survived' column, representing the survival rate
df.survived.mean()
```

```python
# calculates the average survival rate for each gender group
df.groupby('sex').survived.mean()
```

```python
# calculates the average survival rate for each combination of 'sex' and 'class', and displays classes as separate columns
df.groupby(['sex','class']).survived.mean().unstack()
```

### 20. üóëÔ∏è Continous to Categoraical conversion
Binning numeric data into labeled ranges (categories).

```python
# categorizes 'age' into bins: 0-18 as 'childs', 19-25 as 'youngadult', 26-99 as 'adult', and shows the first 5 entries
pd.cut(df.age,bins=[0,18,25,99],labels=['childs','youngadult','adult']).head()
```

```python
# creates a new column 'new-age' by categorizing 'age' into 'childs', 'youngadult', 'adult' for the first 5 rows
df['new-age']=pd.cut(df.age,bins=[0,18,25,99],labels=['childs','youngadult','adult']).head()
df.head()
```

<a id="section21"></a>
### 21. üîÑ Converting one set of values into another one
Mapping or factorizing categorical data into numerical labels.

```python
# creates a new numeric column 'sex_num' mapping 'male' to 0 and 'female' to 1
df['sex_num']=df.sex.map({'male':0,'female':1})
# returns an array of all unique values in the 'embarked' column
df.embarked.unique()
```

```python
# creates a new numeric column 'embarked_num' by assigning a unique integer to each unique value in 'embarked'
df['embarked_num']=df.embarked.factorize()[0]
df.head(5)
```

### 23. üß¨ Reshaping a dataframe
Using the melt function to transform data into a long format, ideal for visualization.

```python
distance=pd.DataFrame([['12345',100,200,300],['34567',400,500,600],['67890',700,800,900]],
                      columns=['zip','factory','warehouse','retail'])
# reshapes the DataFrame from wide to long format with 'zip' as identifier, 'location_type' as variable name, and 'distance1' as values
distance.melt(id_vars='zip',var_name='location_type',value_name='distance1')
import seaborn as sns
sns.barplot(x='zip',y='warehouse',hue='factory',data=distance)
```

```python
distance['factory'] = distance['factory'].astype(str)
sns.barplot(x='zip', y='warehouse', hue='factory', data=distance)
```

```python
sns.barplot(x='zip', y='warehouse', data=distance)
```
